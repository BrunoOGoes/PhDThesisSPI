```python
Nfock = 5

rho_fock = fock_dm(Nfock, 1)
rho_coherent = coherent_dm(Nfock, 1)
```

```python
fig, axes = plt.subplots(1, 2, figsize = (12,6))

plot_fock_distribution(rho_coherent, 
                       fig = fig, 
                       ax = axes[0],
                      title = r"Coherent state, $\beta = 1$")

plot_fock_distribution(rho_fock, 
                       fig = fig, 
                       ax = axes[1],
                      title = r"Fock state, $p_H=1$")

fig.tight_layout()
save_plot('FockOccupation1.jpg')
plt.show()
```

```python
rho_fock2 = (fock_dm(Nfock, 1) + fock_dm(Nfock, 0))/2
rho_coherent2 = coherent_dm(Nfock, 1/2)
```

```python
fig, axes = plt.subplots(1, 2, figsize = (12,6))

plot_fock_distribution(rho_coherent2, 
                       fig = fig, 
                       ax = axes[0],
                      title = r"Coherent state, $\beta = 1/2$")

plot_fock_distribution(rho_fock2, 
                       fig = fig, 
                       ax = axes[1],
                      title = r"Fock state, $p_H=1/2$")

fig.tight_layout()
save_plot('FockOccupation1half.jpg')
plt.show()
```
