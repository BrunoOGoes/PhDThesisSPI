## 1) Spontaneous emission - no spin

\begin{equation}
\left|\Psi(0)\right\rangle =C_{g}\left|g,0\right\rangle +C_{e}\left|e,0\right\rangle 
\end{equation}

\begin{equation}
\left|\Psi(t_{N})\right\rangle =\;C_{g}\left|g,0\right\rangle +C_{e}\left[e^{-\frac{\gamma t_{N}}{2}}\left|e,0\right\rangle -\sqrt{\gamma\Delta t}\sum_{j=0}^{N-1}e^{-i\omega_{0}t_{j}}\left|g,1_{j}\right\rangle \right].
\end{equation}
 

Computing the output field, using $C_{g}=\cos\left(\theta/2\right),C_{e}=\sin\left(\theta/2\right)$

\begin{align}
\langle b_{\text{out}}\rangle & =\left\langle \Psi(t_{N})\right|b_{\text{out}}\left|\Psi(t_{N})\right\rangle \\
 & =-\sqrt{\gamma}\sin\left(\frac{\theta}{2}\right)\cos\left(\frac{\theta}{2}\right)e^{-\frac{\gamma t_{N}}{2}}e^{-i\omega_{0}t_{N}}
\end{align}
 Computing the mean of $\sigma$, in the interaction picture

\begin{align}
\langle\sigma(t_{N})\rangle & =e^{-i\omega_{0}t_{N}}\left\langle \Psi(t_{N})\right|\sigma(t_{N})\left|\Psi(t_{N})\right\rangle \\
 & =e^{-i\omega_{0}t_{N}}\sin\left(\frac{\theta}{2}\right)\cos\left(\frac{\theta}{2}\right)e^{-\frac{\gamma t_{N}}{2}}
\end{align}
 Hence, they are related by,

\begin{equation}
\langle b_{\text{out}}(t_{N})\rangle=-\sqrt{\gamma}\langle\sigma(t_{N})\rangle
\end{equation}

```python
#Parameters
Ω = 1.0
Δω = 0.5*Ω
Γ_pd = 0.0*Ω
Γ = 0.1*Ω
γ = Γ
t_range = np.linspace(0,8*(2+1/8)*np.pi,4*101)
discrete_t_range = np.arange(0,8*(2+1/8)*np.pi,4*np.pi*(1/8))

#Hamiltonian
H_SE = Δω*sigmap()*sigmam()

#Initial state
θ = np.pi/2
ψ0 = np.cos(θ/2)*ket('1') + np.sin(θ/2)*ket('0')

#Collapse operators
c_ops = [np.sqrt(γ)*sigmam(), np.sqrt(Γ_pd)*sigmap()*sigmam()]

expec_list = [sigmap()*sigmam() , sigmam()*sigmap(), sigmam(), sigmax(),sigmay(),sigmaz()]

#Numerical solution of the dynamics
SE_dynamics = mesolve(H_SE, ψ0, t_range, c_ops, expec_list)
```

```python
P_e = SE_dynamics.expect[0]
P_g = SE_dynamics.expect[1]
mean_sm = SE_dynamics.expect[2]
mean_sx = SE_dynamics.expect[3]
mean_sy = SE_dynamics.expect[4]
mean_sz = SE_dynamics.expect[5]
```

```python
####################################
#
#    Collisional model analytical
#
####################################

def Pe_SE(coherence_angle, gamma, t):
    
    return np.sin(coherence_angle/2)**2*np.exp(-gamma*t)

def mean_sigmaSE(Ω0, coherence_angle, γ, t):
    
    value = np.exp(-1j*Ω0*t)*np.sin(coherence_angle/2)*np.cos(coherence_angle/2)*np.exp(-γ*0.5*t)
    
    return np.real(value)
```

```python
discrete_PeSE = []
discrete_cohSE = []

for i in discrete_t_range:
    np.array([discrete_PeSE.append(Pe_SE(θ, γ, i))])
    np.array([discrete_cohSE.append(mean_sigmaSE(Ω, θ, γ, i))])
```

## 1) Spontaneous emission - with spin

We consider the superposition,
\begin{equation}
|\Psi(0)\rangle=\left[C_{\downarrow}|e,\downarrow\rangle+C_{\uparrow}|e,\uparrow\rangle\right]|0_{L},0_{R}\rangle,
\end{equation}
where $|C_{\downarrow}|^{2}+|C_{\uparrow}|^{2}=1$, for convinience
$C_{\downarrow}=\cos\left(\theta/2\right),C_{\uparrow}=\sin\left(\theta/2\right)$.

The state at time $t_{N}$,
\begin{align}
|\Psi(t_{N})\rangle & =e^{-\frac{\eta}{2}N}\left(C_{\downarrow}\left|e,\downarrow,0_{L},0_{R}\right\rangle +C_{\uparrow}\left|e,\uparrow,0_{L},0_{R}\right\rangle \right)\nonumber\\
 & \hspace{1em}-\sqrt{1-e^{-\eta}}\sum_{j=0}^{N-1}e^{-\frac{\eta}{2}j}e^{-i\omega_{0}t_{j}}\left(C_{\downarrow}\left|g,\downarrow,1_{L,(j)},0_{R}\right\rangle +C_{\uparrow}\left|g,\uparrow,0_{L},1_{R,(j)}\right\rangle \right),\nonumber\\
 & =e^{-\frac{\gamma t_{N}}{2}}\left(C_{\downarrow}\left|e,\downarrow,0_{L},0_{R}\right\rangle +C_{\uparrow}\left|e,\uparrow,0_{L},0_{R}\right\rangle \right)\\
 & \hspace{1em}-\sqrt{\Delta t\gamma}\sum_{j=0}^{N-1}e^{-\frac{\gamma}{2}t_{j}}e^{-i\omega_{0}t_{j}}\left(C_{\downarrow}b_{(L)j}^{\dagger}\left|g,\downarrow,0_{L},0_{R}\right\rangle +C_{\uparrow}b_{(R)n}^{\dagger}\left|g,\uparrow,0_{L},0_{R}\right\rangle \right).\nonumber
\end{align}

The probability of being in the excited state is,
\begin{align}
P_{e\uparrow} & =\langle\sigma_{R}^{\dagger}\sigma_{R}\rangle=e^{-\gamma t_{N}}|C_{\uparrow}|^{2}\\
P_{e\downarrow} & =\langle\sigma_{L}^{\dagger}\sigma_{L}\rangle=e^{-\gamma t_{N}}|C_{\downarrow}|^{2}\\
P_{e} & =P_{e\downarrow}+P_{e\uparrow}=e^{-\gamma t_{N}}
\end{align}

The input-output relations are as follow,

\begin{align}
\langle b_{\text{out}}^{L}(t_{N})\rangle & =-\sqrt{\gamma}\langle\sigma_{L}(t_{N})\rangle\\
\langle b_{\text{out}}^{R}(t_{N})\rangle & =-\sqrt{\gamma}\langle\sigma_{R}(t_{N})\rangle
\end{align}

```python
# New Hamiltonian
H_with_spin = Δω*(σ_Rd*σ_R + σ_Ld*σ_L)

# Initial state
θ = np.pi/2
ψ0rl = np.cos(θ/2)*tensor(ket('0'),ket('1')) + np.sin(θ/2)*tensor(ket('0'),ket('0'))

# Collapse operators
c_ops = [np.sqrt(γ)*σ_R, np.sqrt(γ)*σ_L]

# Atomic and spin Pauli operators
#These operators act only on the atom state
sigma_x_at = tensor(sigmax(),qeye(2))
sigma_y_at = tensor(sigmay(),qeye(2))
sigma_z_at = tensor(sigmaz(),qeye(2))

#These operators act only on the spin state
sigma_x_spin = tensor(qeye(2),sigmax())
sigma_y_spin = tensor(qeye(2),sigmay())
sigma_z_spin = tensor(qeye(2),sigmaz())

expec_list = [σ_Rd*σ_R, σ_Ld*σ_L, σ_Rd*σ_R + σ_Ld*σ_L, sigma_x_at, sigma_y_at, sigma_z_at, sigma_x_spin, sigma_y_spin, sigma_z_spin]

result_with_spin = mesolve(H_with_spin, ψ0rl, t_range, c_ops, expec_list)
```

```python
def Pe_SE_Tot(coherence_angle, gamma, t):
    return (np.sin(coherence_angle/2)**2+np.cos(coherence_angle/2)**2)*np.exp(-gamma*t)
```

```python
discrete_PeSE_Tot = []

for i in discrete_t_range:
    np.array([discrete_PeSE_Tot.append(Pe_SE_Tot(θ, γ, i))])
```

```python
mean_σRdσR = result_with_spin.expect[0]
mean_σLdσL = result_with_spin.expect[1]
Pe = result_with_spin.expect[2]


fig, axs = plt.subplots(1,1, figsize = (12,10))


axs.plot(t_range, mean_σRdσR,  label = r"$P_{e\uparrow}$")
axs.plot(t_range, mean_σLdσL,'--',  label = r"$P_{e\downarrow}$")
axs.plot(t_range, Pe,  label = r"$P_e$")
axs.plot(discrete_t_range, discrete_PeSE_Tot, 'o', label = r"$P_e$: CM")

axs.legend(fontsize = 22)
axs.tick_params(axis='x', labelsize=22)#These lines set the font size of the xy labels
axs.tick_params(axis='y', labelsize=22)
axs.set_xlabel(r'$t$', size = 22)
#axs.set_ylabel(r'$\langle\sigma^\dagger \sigma\rangle$', size = 22)
#axs.set_title('Comparison CM vs. numerical solution', size = 22)

#plt.savefig('Comparison-num-vs-analy-Res-Spontaneous-E-with-spin.png',bbox_inches='tight') #Save sempre antes do show!
plt.show()
```
